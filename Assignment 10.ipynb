```python
import requests
from bs4 import BeautifulSoup

def fetch_hyperlinks(url):
    try:
        # Send an HTTP request to the URL
        response = requests.get(url)
        response.raise_for_status()  # Raise an exception for bad responses

        # Parse HTML content using BeautifulSoup
        soup = BeautifulSoup(response.text, 'html.parser')

        # Find all the 'a' tags (hyperlinks) in the HTML
        links = soup.find_all('a')

        # Extract and print the href attribute from each link
        for link in links:
            href = link.get('href')
            if href:
                print(href)

    except requests.exceptions.RequestException as e:
        print(f"Error fetching hyperlinks: {e}")

# Get user input for the website URL
url = input("Enter the website URL: ")

# Call the function to fetch hyperlinks
fetch_hyperlinks(url)
```

3.3.Print all the tags that start from b

```python
from bs4 import BeautifulSoup

# HTML content from the provided page
html_content = """
<html>
<head>
    <title>Page title</title>
</head>
<body>
    <p id="firstpara" align="center">This is paragraph<b>one</b></p>
    <p id="secondpara" align="blah">This is paragraph<b>two</b></p>
</body>
</html>
"""

# Parse HTML content using BeautifulSoup
soup = BeautifulSoup(html_content, 'html.parser')

# Find all tags that start with 'b'
b_start_tags = soup.find_all(lambda tag: tag.name and tag.name.startswith('b'))

# Print the found tags
for tag in b_start_tags:
    print(tag)
```

3.5.Print text from the tags having title and p by using dictionaries

```python
from bs4 import BeautifulSoup

# HTML content from the provided page
html_content = """
<html>
<head>
    <title>Page title</title>
</head>
<body>
    <p id="firstpara" align="center">This is paragraph<b>one</b></p>
    <p id="secondpara" align="blah">This is paragraph<b>two</b></p>
</body>
</html>
"""

# Parse HTML content using BeautifulSoup
soup = BeautifulSoup(html_content, 'html.parser')

# Find all tags with 'title' and 'p'
title_tags = soup.find_all('title')
p_tags = soup.find_all('p')

# Create dictionaries to associate tags with their text content
title_dict = {'title': [tag.text for tag in title_tags]}
p_dict = {'p': [tag.text for tag in p_tags]}

# Print the dictionaries
print("Text from 'title' tags:", title_dict)
print("Text from 'p' tags:", p_dict)
```

3.7.Print the complete tag that hastwo, and only two, attributes

```python
from bs4 import BeautifulSoup

# HTML content from the provided page
html_content = """
<html>
<head>
    <title>Page title</title>
</head>
<body>
    <p id="firstpara" align="center">This is paragraph<b>one</b></p>
    <p id="secondpara" align="blah" class="example">This is paragraph<b>two</b></p>
    <div class="example" data-type="info">Another tag with two attributes</div>
</body>
</html>
"""

# Parse HTML content using BeautifulSoup
soup = BeautifulSoup(html_content, 'html.parser')

# Find tags with exactly two attributes and print the complete tags
tags_with_two_attributes = [tag for tag in soup.find_all() if len(tag.attrs) == 2]

for tag in tags_with_two_attributes:
    print(tag)
```

3.8. Print the tags that have one-character names and no attributes

```python
from bs4 import BeautifulSoup

# HTML content from the provided page
html_content = """
<html>
<head>
    <title>Page title</title>
</head>
<body>
    <p id="firstpara" align="center">This is paragraph<b>one</b></p>
    <a>Link</a>
    <b>Text in bold</b>
</body>
</html>
"""

# Parse HTML content using BeautifulSoup
soup = BeautifulSoup(html_content, 'html.parser')

# Find tags with one-character names and no attributes, and print the complete tags
tags_with_one_char_no_attributes = [tag for tag in soup.find_all() if len(tag.name) == 1 and not tag.attrs]

for tag in tags_with_one_char_no_attributes:
    print(tag)
```

3.9.Print all the tags which have a value of center for align attribute

```python
from bs4 import BeautifulSoup

# HTML content from the provided page
html_content = """
<html>
<head>
    <title>Page title</title>
</head>
<body>
    <p id="firstpara" align="center">This is paragraph<b>one</b></p>
    <p id="secondpara" align="blah">This is paragraph<b>two</b></p>
    <div align="center">This is a div with align attribute set to center</div>
    <span align="center">This is a span with align attribute set to center</span>
</body>
</html>
"""

# Parse HTML content using BeautifulSoup
soup = BeautifulSoup(html_content, 'html.parser')

# Find tags with align attribute set to "center" and print the complete tags
tags_with_center_align = soup.find_all(lambda tag: tag.has_attr('align') and tag['align'] == 'center')

for tag in tags_with_center_align:
    print(tag)
```

3.10.From the XMLcontent'<person name="Bob"><parent rel="mother" name="Alice">'print the attributes having the nameas Alice


```python
import xml.etree.ElementTree as ET

# XML content
xml_content = '<person name="Bob"><parent rel="mother" name="Alice"></parent></person>'

# Parse XML content
root = ET.fromstring(xml_content)

# Find elements with the name attribute set to "Alice" and print their attributes
for element in root.iter():
    if 'name' in element.attrib and element.attrib['name'] == 'Alice':
        print(element.attrib)
```
