3.Plot a pie-chart of the number of models released by every manufacturer, recorded in the provided data. Also,mention the name of the manufacturerwith the largest releases.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
file_path = r"/Users/aksavxavier/Desktop/Datasets/Cars2015.csv"
df = pd.read_csv(file_path)

# Count the number of models released by each manufacturer
manufacturer_counts = df['Make'].value_counts()

# Plot a pie chart
plt.figure(figsize=(10, 8))
plt.pie(manufacturer_counts, labels=manufacturer_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Number of Models Released by Manufacturer')
plt.show()

# Find the manufacturer with the largest number of releases
largest_manufacturer = manufacturer_counts.idxmax()
print(f"The manufacturer with the largest number of releases is: {largest_manufacturer}")
```

```python
import matplotlib.pyplot as plt

# Given data points
X = [1, 2, 3, 4]
Y = [20, 21, 20.5, 20.8]

# Error values
y_error = [0.12, 0.13, 0.2, 0.1]

# Define figure size and adjust plot DPI
fig, ax = plt.subplots(figsize=(4, 5), dpi=100)

# Draw a simple plot with configured axes and labels
ax.errorbar(X, Y, yerr=y_error, linestyle='--', marker='s', color='b', label='Data Points with Error Bars')

# Configure the X-axis
ax.set_xlim(0, 5)  # Set X-axis limits
ax.set_xticks([1, 2, 3, 4, 5])  # Set X-axis tick locations
ax.set_xlabel('X-axis', fontsize=14)  # Set X-axis label with font size

# Configure the Y-axis
ax.set_ylim(19.5, 21.5)  # Set Y-axis limits
ax.set_yticks([19.5, 20, 20.5, 21, 21.5])  # Set Y-axis tick locations
ax.set_ylabel('Y-axis', fontsize=14)  # Set Y-axis label with font size

# Add a grid
ax.grid(True, linestyle='--', alpha=0.7)

# Add title and labels with font size
ax.set_title('Configured Plot with Error Bars', fontsize=14)  # Set the title with font size

# Add legend with font size
ax.legend(fontsize=14)

# Set tick label font size
ax.tick_params(axis='both', labelsize=14)

# Show the plot
plt.show()
```

1.Plot total sales per month fortheYear 2011.  How have the total sales increased over the months in the Year 2011? Which month has the lowest sales

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
file_path =r"/Users/aksavxavier/Desktop/Datasets/BigMartSalesData_bo90s.csv"
df = pd.read_csv(file_path)

# Convert the 'InvoiceDate' column to datetime type
df['InvoiceDate'] = pd.to_datetime(df['InvoiceDate'])

# Extract the month and year from the 'InvoiceDate' column
df['YearMonth'] = df['InvoiceDate'].dt.to_period('M')

# Group by month and sum the 'Amount' column to get total sales per month
monthly_sales = df.groupby('YearMonth')['Amount'].sum()

# Plot the total sales per month
plt.figure(figsize=(10, 6))
monthly_sales.plot(kind='bar', color='skyblue')
plt.title('Total Sales per Month in 2011')
plt.xlabel('Month')
plt.ylabel('Total Sales')
plt.show()
```

2.Plot a Pie Chart for the Year 2011 Country Wise. Which Country contributes highest towards sales?

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
file_path =r"/Users/aksavxavier/Desktop/Datasets/BigMartSalesData_bo90s.csv"
df = pd.read_csv(file_path)

# Convert the 'InvoiceDate' column to datetime type
df['InvoiceDate'] = pd.to_datetime(df['InvoiceDate'])

# Extract the year from the 'InvoiceDate' column
df['Year'] = df['InvoiceDate'].dt.year

# Filter data for the Year 2011
df_2011 = df[df['Year'] == 2011]

# Group by country and sum the 'Amount' column to get total sales by country
country_sales = df_2011.groupby('Country')['Amount'].sum()

# Plot a pie chart for sales by country
plt.figure(figsize=(10, 8))
country_sales.plot(kind='pie', autopct='%1.1f%%', startangle=90, cmap='viridis')
plt.title('Sales Distribution by Country in 2011')
plt.ylabel('')  # Remove y-axis label
plt.show()
```
